构建模型（随机森林）

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
import matplotlib
```

```python
# 添加常数项X0 = 1，用于偏置
X_train_b = np.c_[np.ones((X_train_scaled.shape[0], 1)), X_train_scaled]  # 训练集添加x0 = 1
X_test_b = np.c_[np.ones((X_test_scaled.shape[0], 1)), X_test_scaled]  # 测试集添加x0 = 1
```

```python
# 使用牛顿法求解线性回归，带有正则化项来避免 Hessian 矩阵不可逆
def newton_method(X, y, tolerance=1e-6, max_iterations=100, lambda_reg=1e-4):
    m, n = X.shape  # 样本数和特征数
    theta = np.random.randn(n, 1)  # 初始化参数

    for iteration in range(max_iterations):
        # 计算预测值
        predictions = X.dot(theta)

        # 计算梯度
        gradients = X.T.dot(predictions - y) / m

        # 计算Hessian矩阵
        hessian = X.T.dot(X) / m

        # 引入正则化项以避免 Hessian 矩阵不可逆的情况
        hessian += lambda_reg * np.eye(n)

        # 计算参数更新量
        try:
            delta_theta = np.linalg.inv(hessian).dot(gradients)
        except np.linalg.LinAlgError:
            print("Hessian矩阵不可逆，尝试使用不同的正则化参数")
            return theta

        # 更新参数
        theta = theta - delta_theta

        # 如果参数更新量足够小，停止迭代
        if np.linalg.norm(delta_theta, ord=2) < tolerance:
            print(f'在 {iteration + 1} 次迭代后收敛')
            break

    return theta
```

```python
# 使用牛顿法进行线性回归
theta_newton = newton_method(X_train_b, y_train.values.reshape(-1, 1))
```

```python
# 训练集预测
y_pred_train_newton = X_train_b.dot(theta_newton)
```

```python
# 测试集预测
y_pred_test_newton = X_test_b.dot(theta_newton)
```

```python
# 评估模型性能
mse_newton = mean_squared_error(y_test, y_pred_test_newton)
r2_newton = r2_score(y_test, y_pred_test_newton)
mae_newton = mean_absolute_error(y_test, y_pred_test_newton)

print(f'\n牛顿法线性回归 - 均方误差 (MSE): {mse_newton}')
print(f'牛顿法线性回归 - R² 评分: {r2_newton}')
print(f'牛顿法线性回归 - 平均绝对误差 (MAE): {mae_newton}')
```

```python
# 5. 残差分析

# 残差计算
residuals_newton = y_test.values.reshape(-1, 1) - y_pred_test_newton

# 如果数据量不大，可以直接使用所有残差
residuals_sample = residuals_newton.ravel()

plt.figure(figsize=(10, 6))
plt.hist(residuals_sample, bins=50, color='blue', alpha=0.7)
plt.title('牛顿法线性回归残差分布')
plt.savefig('result2_with_handwritten_newton/residuals_hist_newton.png')  # 保存图像到 result2 文件夹
plt.close()  # 释放内存
```
